```python
"""
Exploratory Data Analysis for Heart Disease UCI Dataset
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create output directory for plots
os.makedirs("screenshots", exist_ok=True)

# ============================================================
# 1. LOAD DATA
# ============================================================
print("=" * 60)
print("HEART DISEASE UCI - EXPLORATORY DATA ANALYSIS")
print("=" * 60)

df = pd.read_csv("data/heart.csv")

print("\n--- Dataset Shape ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n--- First 5 Rows ---")
print(df.head())

print("\n--- Data Types ---")
print(df.dtypes)

print("\n--- Statistical Summary ---")
print(df.describe())

print("\n--- Missing Values ---")
print(df.isnull().sum())

# ============================================================
# 2. HANDLE MISSING VALUES
# ============================================================
print("\n--- Handling Missing Values ---")
# Fill missing values with median for numerical columns
for col in df.columns:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"  Filled {col} missing values with median: {median_val}")

print(f"\nMissing values after cleaning: {df.isnull().sum().sum()}")

# Save cleaned data
df.to_csv("data/heart_cleaned.csv", index=False)
print("Cleaned data saved to data/heart_cleaned.csv")

# ============================================================
# 3. VISUALIZATIONS
# ============================================================

# --- Plot 1: Target Distribution (Class Balance) ---
fig, ax = plt.subplots(figsize=(8, 5))
target_counts = df['target'].value_counts()
colors = ['#2ecc71', '#e74c3c']
bars = ax.bar(target_counts.index, target_counts.values, color=colors, edgecolor='black')
ax.set_xlabel('Target (0 = No Disease, 1 = Disease)', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Heart Disease - Class Distribution', fontsize=14, fontweight='bold')
ax.set_xticks([0, 1])
ax.set_xticklabels(['No Disease (0)', 'Disease (1)'])
for bar, count in zip(bars, target_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 2,
            str(count), ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.savefig("screenshots/01_class_distribution.png", dpi=150)
plt.close()
print("Saved: screenshots/01_class_distribution.png")

# --- Plot 2: Histograms of All Features ---
fig, axes = plt.subplots(4, 4, figsize=(16, 12))
axes = axes.flatten()
for i, col in enumerate(df.columns):
    if i < len(axes):
        df[col].hist(ax=axes[i], bins=20, color='steelblue', edgecolor='black', alpha=0.7)
        axes[i].set_title(col, fontsize=10, fontweight='bold')
        axes[i].tick_params(labelsize=8)
# Hide empty subplots
for j in range(len(df.columns), len(axes)):
    axes[j].set_visible(False)
fig.suptitle('Feature Distributions', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig("screenshots/02_feature_histograms.png", dpi=150)
plt.close()
print("Saved: screenshots/02_feature_histograms.png")

# --- Plot 3: Correlation Heatmap ---
fig, ax = plt.subplots(figsize=(12, 10))
corr_matrix = df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, linewidths=0.5, ax=ax,
            cbar_kws={"shrink": 0.8})
ax.set_title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig("screenshots/03_correlation_heatmap.png", dpi=150)
plt.close()
print("Saved: screenshots/03_correlation_heatmap.png")

# --- Plot 4: Age Distribution by Target ---
fig, ax = plt.subplots(figsize=(10, 6))
df[df['target'] == 0]['age'].hist(ax=ax, bins=20, alpha=0.6, label='No Disease', color='#2ecc71', edgecolor='black')
df[df['target'] == 1]['age'].hist(ax=ax, bins=20, alpha=0.6, label='Disease', color='#e74c3c', edgecolor='black')
ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Age Distribution by Heart Disease Status', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
plt.tight_layout()
plt.savefig("screenshots/04_age_distribution.png", dpi=150)
plt.close()
print("Saved: screenshots/04_age_distribution.png")

# --- Plot 5: Box plots of key features by target ---
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
key_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca']
for i, feat in enumerate(key_features):
    row, col_idx = i // 3, i % 3
    sns.boxplot(x='target', y=feat, data=df, ax=axes[row][col_idx], 
                palette=['#2ecc71', '#e74c3c'])
    axes[row][col_idx].set_title(f'{feat} by Target', fontweight='bold')
    axes[row][col_idx].set_xticklabels(['No Disease', 'Disease'])
fig.suptitle('Key Features by Heart Disease Status', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("screenshots/05_boxplots.png", dpi=150)
plt.close()
print("Saved: screenshots/05_boxplots.png")

print("\n✅ EDA Complete! All visualizations saved in screenshots/ folder.")
```
